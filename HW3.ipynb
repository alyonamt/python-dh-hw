# *Задание №1*
1. Скачайте текст "Литературных анекдотов". 
2. Напишите функцию, которая будет читать файл, лемматизировать текст с помощью pymystem3 и записывать результат в новый файл. У функции должно бы два аргумента: путь к исходному файлу и путь к файлу с лемматизированным текстом. Вызов функции тоже должен быть прописан в решении.

```python
from pymystem3 import Mystem
m = Mystem()
```

```python
def lemmatize(input_path, output_path):
    with open(input_path, 'r', encoding='utf-8') as f:
        text = f.read()
    
    lemmas = m.lemmatize(text)
    with open(output_path, 'w', encoding='utf-8') as g:
        g.write(''.join(lemmas))
```

```python
lemmatize('литературные анекдоты.txt', 'лемматизированные анекдоты.txt')
```

# *Задание №2*
1. Очистите лемматизированный текст от стоп-слов и посчитайте ipm для оставшихся. 
2. Выведите 20 самых частотных по этому параметру слов.

```python
from collections import Counter

def task2(path):
    with open(path, 'r', encoding='utf-8') as f:
        text = f.read()
        
    words = [word.strip("()!?.,-\"") for word in text.split()]
    
    with open('rus_stopwords.txt', 'r', encoding='utf-8') as g:
        stopwords = g.read().split() + [''] 
        
    no_stopwords = [w for w in words if w not in stopwords]
    most_common = Counter(no_stopwords).most_common(20)
    
    count_words = len(no_stopwords)
    most_commons_ipm = [(w, (c / count_words) * 1000000) for w, c in most_common]
    print(*most_commons_ipm, sep='\n')
```

```python
task2('лемматизированные анекдоты.txt')
```

# *Задание №3*
1. Сделайте полный морфологический разбор исходного текста. 
2. Напишите регулярное выражение, которое будет извлекать из тега только часть речи. 
3. Пройдитесь циклом по списку с разборами, который выдал pymystem3, извлекая из каждого разбора форму слова и его часть речи и записывая их в новый словарь (форма -- ключ, часть речи -- значение). 
4. Посчитайте абсолютную частоту для всех частей речи, а затем относительнную частоту (абсолютная частота / длина текста в словах).

```python
with open('литературные анекдоты.txt', 'r', encoding='utf-8') as f:
    text = f.read()
```

```python
# 1
gr_text = m.analyze(text)
```

```python
# 2
from re import compile


part_of_speech = compile('[A-Z]+')

parts_of_speech = part_of_speech.findall(str(gr_text))
print(parts_of_speech)
```

```python
# 3
words_speech = {}
for word in gr_text:
    try:
        form = word['text']
        gr = part_of_speech.match(word['analysis'][0]['gr']).group()
        words_speech[form] = gr
    except (KeyError, IndexError) as e:
        pass
print(words_speech)
```

```python
# 4
from collections import Counter

abs_frequency = list(Counter(parts_of_speech).items())
# * и \n здесь для красоты
print('Absolute Frequency')
print(*abs_frequency, sep='\n')

count = len(parts_of_speech)

rel_frequency = [(w, c / count) for w, c in abs_frequency]
print('\nRelative Frequency')
print(*rel_frequency, sep='\n')
```
